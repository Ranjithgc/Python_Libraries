***@Title : Program Aim to draw a boxplot of day against tip.***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style('whitegrid')
# Import Data
df = sns.load_dataset('tips')

# Construct plot
sns.boxplot(x="tip", y="day", hue="smoker",
                data=df, palette="Set2", orient = "h")
plt.show()
```
