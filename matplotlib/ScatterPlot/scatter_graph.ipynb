***@Title : Program Aim to draw a scatter graph taking a random distribution in X and Y and plotted against each other.***

```python
import matplotlib.pyplot as plt
from loggers import logger
from pylab import randn

try:
    X = randn(200)
    Y = randn(200)
    plt.scatter(X,Y, color='r')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
except Exception as e:
    logger.error(e)
```
