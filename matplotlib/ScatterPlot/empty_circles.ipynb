***@Title : Program Aim to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.***

```python
import matplotlib.pyplot as plt 
import numpy as np 
from loggers import logger

try:
    x = np.random.randn(50) 
    y = np.random.randn(50)
    plt.scatter(x, y, s=70, facecolors='none', edgecolors='g')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
except Exception as e:
    logger.error(e)
```
