***@Title : Program Aim to create a pie chart of the popularity of programming Languages.***

```python
import matplotlib.pyplot as plt
from loggers import logger

try:
    # Data to plot
    languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
    popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
        
    # Plot
    plt.pie(popuratity, labels=languages, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=90)

    plt.axis('equal')
    plt.show()
except Exception as e:
    logger.error(e)
```

***2. Add title to the pie chart.***

```python
plt.pie(popuratity, labels=languages, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago", bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```

***3. make multiple wedges to the pie.***

```python
# explode 1st slice
explode = (0.1, 0, 0, 0,0,0)  
    
plt.pie(popuratity, explode=explode, labels=languages, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago", bbox={'facecolor':'0.8', 'pad':5})
plt.legend(title = "languages")
plt.show()
```
